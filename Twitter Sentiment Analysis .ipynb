```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.cluster import KMeans
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
df1['Target'].unique()
```

```python
plt.pie(df1['Target'].value_counts(), labels=df1['Target'].unique(), autopct='%1.1f%%',
            startangle=90, wedgeprops={'linewidth': 0.5}, textprops={'fontsize': 12},
            explode=[0.1, 0.1, 0.1, 0.1], shadow=True)
plt.show()
```

```python
# Filter the DataFrame to include only 'Negative' Target
negative_df1 = df1[df1['Target'] == 'Negative']

# Create the plot
plt.figure(figsize=(20, 7))
sns.countplot(x="Platform", hue="Target", data=negative_df1,palette='flare')
plt.title("Platform by Negative Target")
plt.xticks(rotation=90)
plt.show()
```

```python
# Filter the DataFrame to include only 'Positive' Target
negative_df1 = df1[df1['Target'] == 'Positive']

# Create the plot
plt.figure(figsize=(20, 7))
sns.countplot(x="Platform", hue="Target", data=negative_df1,palette='flare')
plt.title("Platform by positive Target")
plt.xticks(rotation=90)
plt.show()
```

```python
# Create cross-tabulation table in plot
plt.figure(figsize=(10, 9))
count_table = pd.crosstab(index=df1['Platform'], columns=df1['Target'])
sns.heatmap(count_table, cmap='RdPu', annot=True, fmt='d',linewidths=0.5, linecolor='black')
plt.title('Target Distribution by Platform')
plt.xlabel('Target')
plt.ylabel('Platform')
plt.show()
```
